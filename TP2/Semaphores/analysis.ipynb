```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(context = 'paper', style='whitegrid', font_scale = 1.4)
```

```
path = 'report.csv'
data = pd.read_csv(path, names=['N','Np','Nc','Tempo'])
data['Np,Nc'] = data[['Np','Nc']].apply(tuple, axis=1)
data
```

```
fig = plt.figure(figsize=(12,6))
fig.suptitle('Número de threads produtor X Tempo', fontsize=20)
plt.xlabel('Np', fontsize=15)
plt.ylabel('Tempo (s)', fontsize=15)
sns.lineplot(
  data = data,
  x = "Np",
  y = "Tempo",
  hue = "N",
  linewidth = 3,
  palette = ["blue", "red", "green", "yellow", "pink"]
)
plt.show()

fig2 = plt.figure(figsize=(12,6))
fig2.suptitle('Número de threads consumidor X Tempo', fontsize=20)
plt.xlabel('Nc', fontsize=15)
plt.ylabel('Tempo (s)', fontsize=15)
sns.lineplot(
  data = data,
  x = "Nc",
  y = "Tempo",
  hue = "N",
  linewidth = 3,
  palette = ["blue", "red", "green", "yellow", "pink"]
)
plt.show()
```

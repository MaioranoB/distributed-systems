```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(context = 'paper', style='whitegrid', font_scale = 1.4)
```

```python
path = 'report.csv'
data = pd.read_csv(path, names=['N','K','Tempo'])
data
```

```python
fig = plt.figure(figsize=(12,6))
fig.suptitle('NÃºmero de threads X Tempo', fontsize=20)
plt.xlabel('K', fontsize=15)
plt.ylabel('Tempo (s)', fontsize=15)
sns.lineplot(
  data = data,
  x = "K",
  y = "Tempo",
  hue = "N",
  linewidth = 3,
  palette = ["blue", "red", "green"]
)
plt.show()
```
